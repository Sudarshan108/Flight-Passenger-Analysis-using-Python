# Create a new dataframe “df” by joining the two datasets

```python
df = passenger_data.set_index('id').join(survey_data.set_index('id'))
```

```python
df
```

```python
df.isnull()
```

# Deal for missing values

```python
df = df.fillna(df.mean())
```

```python
df
```

# Drop the duplicate data

```python
df = df.drop_duplicates()
```

```python
df
```

# Rename the columns with string title format 

```python
df = df.rename(columns={'Type of Travel': 'Type Of Travel', 
                        'Inflight wifi service': 'Inflight Wifi Service', 
                        'Departure/Arrival time convenient': 'Departure/Arrival Time Convenient',
                        'Ease of Online booking': 'Ease Of Online Booking',
                        'Gate location':'Gate Location',
                        'Food and drink':'Food and drink',
                        'Online boarding':'Online Boarding',
                        'Leg room service':'Leg Room Service',
                        'Baggage handling':'Baggage Handling',
                        'Checkin service':'Checkin Service',
                        'Inflight entertainment':'Inflight Entertainment',
                        'Departure Delay in Minutes':'Departure Delay In Minutes',
                        'Arrival Delay in Minutes':'Arrival Delay In Minutes',
                        'satisfaction':'Satisfaction'})
```

```python
df
```

```python
df.columns
```

# Set “ID” as Index columns

# Replace the Column values in “Satisfaction” column with Logic;

```python
replacement_dict = {'satisfied': 'Yes', 'neutral or dissatisfied': 'No'}
df['Satisfaction'] = df['Satisfaction'].replace(replacement_dict)
```

# To find the hidden insights on the dataframe “df”.
1.Analysis of Satisfaction Levels by Travel Class:

```python

satisfaction_class = df.groupby('Class')['Satisfaction'].value_counts(normalize=True).unstack()
satisfaction_class.plot(kind='bar', stacked=True)
plt.xlabel('Class')
plt.ylabel('Percentage')
plt.title('Satisfaction Levels by Travel Class')
plt.legend(title='Satisfaction')
plt.show()
```

2.Analysis of Satisfaction Levels by Type of Travel:


```python
satisfaction_type = df.groupby('On-board service')['Satisfaction'].value_counts(normalize=True).unstack()
satisfaction_type.plot(kind='bar', stacked=True)
plt.xlabel('On-board service')
plt.ylabel('Percentage')
plt.title('Satisfaction Levels by On-board service')
plt.legend(title='Satisfaction')
plt.show()
```

3.Analysis of Flight Distance and Customer Type:


```python
plt.boxplot([df[df['Class'] == 'Loyal customer']['Flight Distance'],
             df[df['Class'] == 'Disloyal customer']['Flight Distance']],
            labels=['Loyal customer', 'Disloyal customer'])
plt.xlabel('Class')
plt.ylabel('Flight Distance')
plt.title('Flight Distance by Class')
plt.show()
```

4.Analysis of Age Distribution by Gender:

```python
sns.kdeplot(data=df, x='Age', hue='Gender', fill=True)
plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Age Distribution by Gender')
plt.legend(title='Gender')
plt.show()
```

5.Analyzing satisfaction levels by age group:

```python
df['Age Group'] = pd.cut(df['Age'], bins=[0, 18, 30, 45, 60, np.inf], labels=['0-18', '19-30', '31-45', '46-60', '60+'])
satisfaction_age = df.groupby('Age Group')['Satisfaction'].value_counts(normalize=True).unstack()
satisfaction_age.plot(kind='bar', stacked=True)
plt.xlabel('Age Group')
plt.ylabel('Percentage')
plt.title('Satisfaction Levels by Age Group')
plt.legend(title='Satisfaction')
plt.show()
```

6.Analyzing flight distance by travel purpose:

```python
avg_distance_purpose = df.groupby('Type Of Travel')['Flight Distance'].mean()
avg_distance_purpose.plot(kind='bar')
plt.xlabel('Travel Purpose')
plt.ylabel('Average Flight Distance')
plt.title('Average Flight Distance by Travel Purpose')
plt.show()
```

7.Analyzing satisfaction levels by gender and travel purpose:

```python
satisfaction_gender_purpose = df.groupby(['Gender', 'Type Of Travel'])['Satisfaction'].value_counts(normalize=True).unstack()
satisfaction_gender_purpose.plot(kind='bar', stacked=True)
plt.xlabel('Gender and Travel Purpose')
plt.ylabel('Percentage')
plt.title('Satisfaction Levels by Gender and Travel Purpose')
plt.legend(title='Satisfaction')
plt.show()
```

# Select numerical columns


```python
numerical_columns = df.select_dtypes(include='number')
print(numerical_columns)
```


# Select categorical columns




```python
categorical_columns = df.select_dtypes(include='object')
print(categorical_columns)
```

# Create dummy variables for categorical columns



```python
dummy_columns = pd.get_dummies(categorical_columns)
print(dummy_columns)
```

# Concatenate numerical and dummy columns



```python
newdata = pd.concat([numerical_columns, dummy_columns], axis=1)
newdata
```


# Perform correlation analysis


```python
correlation_matrix = newdata.corr()
correlation_matrix 
```

```python
plt.figure(figsize=(20, 10))
sns.heatmap(correlation_matrix, annot=True, linewidths=2)
plt.title('Correlation Heatmap')
plt.show()
```
